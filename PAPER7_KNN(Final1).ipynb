```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/PW4-DataSets/train1.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head(20)
```

```python
df.isnull().mean() * 100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head(20)
```

```python
knn = KNNImputer(n_neighbors=3,weights='distance')

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf,y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

```python
lr = LogisticRegression()
lr.fit(X_train_trf, y_train)
y_pred= lr.predict(X_test_trf)
accuracy = accuracy_score(y_test, y_pred)
accuracy_percentage = accuracy * 100
print(f"Accuracy: {accuracy_percentage:.2f}%")
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```

```python
lr = LogisticRegression()
lr.fit(X_train_trf2, y_train)
y_pred2 = lr.predict(X_test_trf2)
accuracy = accuracy_score(y_test, y_pred2)
accuracy_percentage = accuracy * 100
print(f"Accuracy: {accuracy_percentage:.2f}%")

```

```python





# Create an instance of SimpleImputer with median strategy
si = SimpleImputer(strategy='median')

# Perform median imputation on training data
X_train_trf2 = si.fit_transform(X_train)
# Perform median imputation on test data
X_test_trf2 = si.transform(X_test)

# Create a logistic regression model
lr = LogisticRegression()

# Fit the model on the imputed training data
lr.fit(X_train_trf2, y_train)

# Make predictions on the imputed test data
y_pred2 = lr.predict(X_test_trf2)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred2)
# Convert accuracy to percentage
accuracy_percentage = accuracy * 100

print("Accuracy: {:.2f}%".format(accuracy_percentage))

```

```python

# Create an instance of SimpleImputer with mode strategy
si = SimpleImputer(strategy='most_frequent')

# Perform mode imputation on training data
X_train_trf2 = si.fit_transform(X_train)
# Perform mode imputation on test data
X_test_trf2 = si.transform(X_test)

# Create a logistic regression model
lr = LogisticRegression()

# Fit the model on the imputed training data
lr.fit(X_train_trf2, y_train)

# Make predictions on the imputed test data
y_pred2 = lr.predict(X_test_trf2)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred2)

# Convert accuracy to percentage
accuracy_percentage = accuracy * 100

print("Accuracy: {:.2f}%".format(accuracy_percentage))
```
